## 概念讲解 Concepts

### 为什么需要面向对象？

过程式（procedural）编程把数据和操作分开放；
面向对象把"数据 + 对数据的操作"打包成一个**对象（object）**，让代码更易于扩展和复用。

---

### `class` 与 `__init__`

```python
class Dog:
    def __init__(self, name: str, age: int):
        self.name = name   # 实例属性（instance attribute）
        self.age  = age

    def bark(self) -> str:
        return f"{self.name} says: Woof!"
```

- `__init__` 是**构造方法**（constructor），对象被创建时自动调用。
- `self` 是指向当前对象本身的引用，**第一个参数永远是 self**（约定俗成的名字）。
- `self.name` 存在**实例**上，每个对象各有一份。

---

### 方法（Method）

方法就是定义在类里的函数。调用时 Python 自动把对象本身作为第一个参数传入：

```python
d = Dog("Buddy", 3)
d.bark()          # 等价于 Dog.bark(d)
```

---

### 继承（Inheritance）与 `super()`

```python
class GuideDog(Dog):
    def __init__(self, name: str, age: int, owner: str):
        super().__init__(name, age)   # 先调用父类 __init__
        self.owner = owner

    def bark(self) -> str:           # override 父类方法
        return f"{self.name} (导盲犬) says: Woof!"
```

- `super()` 返回父类对象，常用于在子类中调用父类的 `__init__` 或方法。
- 子类**重新定义**同名方法，叫做**覆盖（override）**；调用时 Python 会找到最近的那个版本。

---

### 多态（Polymorphism）

同一个函数签名，不同的子类有不同的实现。调用者无需关心具体类型：

```python
def make_sound(dog: Dog) -> str:
    return dog.bark()   # 自动调用正确版本
```

---

### `isinstance()`

```python
isinstance(d, Dog)        # True
isinstance(d, GuideDog)   # 取决于 d 的实际类型
```

用于在运行时检查对象是否属于某个类或其子类。

---

### 抽象基类模式（Abstract Base Pattern）

常见做法：在基类中定义方法但 `raise NotImplementedError`，强制子类实现：

```python
class Shape:
    def area(self) -> float:
        raise NotImplementedError("子类必须实现 area()")
```

这不是真正的抽象类（那需要 `abc.ABC`），但对初学者足够用。


## 示例 Worked Example

下面是 `w1-oop` 练习包里的核心层次结构：`Shape` → `Rectangle` / `Circle`。
代码完全自包含，可直接运行。


````python
import math
from typing import List


class Shape:
    """所有形状的基类（Base class）。"""

    def area(self) -> float:
        raise NotImplementedError("子类必须实现 area()")

    def __repr__(self) -> str:
        return f"{type(self).__name__}(area={self.area():.4f})"


class Rectangle(Shape):
    """矩形：宽 w，高 h。"""

    def __init__(self, w: float, h: float):
        self.w = w
        self.h = h

    def area(self) -> float:
        return self.w * self.h


class Circle(Shape):
    """圆形：半径 r。"""

    def __init__(self, r: float):
        self.r = r

    def area(self) -> float:
        return math.pi * self.r ** 2


def total_area(shapes: List[Shape]) -> float:
    """计算所有形状面积之和（多态用法）。"""
    return sum(s.area() for s in shapes)


# ── 快速验证 ──────────────────────────────────────────────────────────────
r = Rectangle(4, 5)
c = Circle(3)

print(r)
print(c)
print(f"total_area = {total_area([r, c]):.4f}")

# isinstance 演示
for shape in [r, c]:
    print(f"  {shape!r}  →  isinstance(Shape)={isinstance(shape, Shape)}")
````
