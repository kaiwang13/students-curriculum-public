## 概念讲解 Concepts

### 为什么要用 NumPy？

Python 原生列表（`list`）是通用对象容器，每次数学运算都需要 Python 解释器逐元素循环。
NumPy 把**同类型数据**存储在连续内存块（C 数组），借助 BLAS/LAPACK 和向量化 CPU 指令，
速度比纯 Python 快 10–1000 倍，且代码更简洁。

---

### 数组创建 Creating arrays

```python
import numpy as np

# 从列表（list）创建
a = np.array([1, 2, 3])          # 1D，shape=(3,)
B = np.array([[1, 2], [3, 4]])   # 2D，shape=(2,2)

# 工厂函数
np.zeros((2, 3))                 # 全 0，shape=(2,3)
np.ones((3,), dtype=np.int32)   # 全 1，int32
np.arange(0, 10, 2)             # [0, 2, 4, 6, 8]
np.linspace(0, 1, 5)            # 等间距 5 点，含端点
```

---

### 形状与维度 Shape & ndim

| 属性 | 含义 |
|------|------|
| `a.shape` | 各维度大小的元组，如 `(3, 4)` |
| `a.ndim`  | 维度数（阶数），如 `2` |
| `a.size`  | 元素总数 = `3 × 4 = 12` |

---

### 数据类型 dtype

NumPy 数组中所有元素类型相同。常见 dtype：

| dtype | Python 等价 | 字节 |
|-------|------------|------|
| `np.float64` | `float` | 8 |
| `np.int32` | `int`（32 位）| 4 |
| `np.bool_` | `bool` | 1 |
| `np.complex128` | `complex` | 16 |

用 `astype` 可以转换（返回新数组，不修改原数组）：

```python
a = np.array([1, 2, 3])        # dtype=int64（默认）
b = a.astype(np.float64)       # [1., 2., 3.]
```

---

### 内存布局 Memory layout

NumPy 默认**行优先（C order）**：`a[i, j]` 与 `a[i, j+1]` 在内存相邻。
Fortran 列优先（F order）用 `order='F'`。大多数情况用默认即可。


## 示例 Worked Example

````python
import numpy as np

# 1. 创建各种数组
a = np.array([10, 20, 30, 40, 50], dtype=np.float32)
B = np.arange(12).reshape(3, 4)   # shape=(3,4)
C = np.linspace(0.0, 1.0, 6).reshape(2, 3)

print("a:", a, "  dtype:", a.dtype, "  shape:", a.shape)
print("B:\n", B, "  shape:", B.shape, "  size:", B.size)
print("C:\n", C)

# 2. dtype 转换
B_float = B.astype(np.float64)
print("\nB_float dtype:", B_float.dtype)

# 3. 零矩阵与单位矩阵
Z = np.zeros((3, 3))
I = np.eye(3)           # 单位矩阵（identity matrix）
print("\nzeros:\n", Z)
print("eye:\n", I)

# 4. 常用属性
print("\nB.shape =", B.shape, "  B.ndim =", B.ndim, "  B.size =", B.size)
````
