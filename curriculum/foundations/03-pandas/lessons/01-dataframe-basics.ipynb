## 概念讲解 Concepts

### Series 与 DataFrame

**Series** 是带标签的一维数组——相当于「一列数据 + 行标签」：

```python
import pandas as pd
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
```

**DataFrame** 是带行、列标签的二维表——由多个 Series 组成：

```python
df = pd.DataFrame({"name": ["Alice", "Bob"], "age": [25, 30]})
```

---

### 常用检视函数 Inspection

| 方法 / 属性 | 作用 |
|-------------|------|
| `df.head(n)` / `df.tail(n)` | 前 / 后 n 行，默认 5 |
| `df.shape` | `(行数, 列数)` 元组 |
| `df.dtypes` | 每列的数据类型 |
| `df.info()` | 类型 + 非空计数 |
| `df.describe()` | 数值列统计摘要 |

---

### 索引：`loc` vs `iloc`

| 索引器 | 用法 | 依据 |
|--------|------|------|
| `df.loc[行标签, 列标签]` | 按**标签**（含端点）| index / column 名 |
| `df.iloc[行整数, 列整数]` | 按**位置**（不含右端点）| 0-based 整数 |

常见用法：
```python
df.loc[0:2, "age"]          # 行标签 0–2，列名 "age"
df.iloc[0:3, 1]             # 位置行 0–2，第 1 列（不含 3）
df.loc[df["age"] > 25]      # 布尔过滤
df[["name", "age"]]         # 多列选取（双括号！）
```

> **注意**：`loc` 区间含右端点，`iloc` 不含。


## 示例 Worked Example

````python
import pandas as pd
import numpy as np

# 1. 创建 DataFrame
data = {
    'record_id': [101, 102, 103, 104, 105],
    'age':       [34, 56, 23, 71, 45],
    'group':     ['A', 'B', 'A', 'B', 'A'],
    'x1':        [22.1, 28.4, 19.7, 31.2, 25.6],
    'label':      [0, 1, 0, 1, 0],
}
df = pd.DataFrame(data)
print('=== head(3) ===')
print(df.head(3))

# 2. 检视结构
print('\nshape:', df.shape)
print('\ndtypes:\n', df.dtypes)

# 3. loc 按标签选取（含右端点 3）
print('\n=== loc: 行 1-3, 列 age/group ===')
print(df.loc[1:3, ['age', 'group']])

# 4. iloc 按位置选取（不含右端点 4）
print('\n=== iloc: 前 2 行，第 2-3 列 ===')
print(df.iloc[:2, 2:4])

# 5. 布尔过滤
old_records = df.loc[df['age'] > 50]
print('\n=== 年龄 > 50 ===')
print(old_records)
````

````python
# TODO：在这里完成动手练习
import pandas as pd

# 1. 创建 DataFrame
records = {
    'name':   ['Alice', 'Bob', 'Carol', 'Dave', 'Eve', 'Frank'],
    'score':  [88, 72, 95, 60, 83, 77],
    'grade':  ['B', 'C', 'A', 'D', 'B', 'C'],
}
df2 = pd.DataFrame(records)

# 2. 检视
print('shape:', df2.shape)
print('dtypes:\n', df2.dtypes)

# 3. iloc 取位置 1-3 行，第 1 列
print('\niloc[1:4, 1]:\n', df2.iloc[1:4, 1])

# 4. 布尔过滤：成绩 >= 80
print('\n成绩 >= 80:\n', df2.loc[df2['score'] >= 80])
````
