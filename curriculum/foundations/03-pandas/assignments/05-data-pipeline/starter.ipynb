## 2. 探索性数据分析（EDA）

在下面两个单元格中分析脏数据的特征：
- 缺失值分布
- 数值列分布

每张图都需要标注轴标签和标题。

````python
# EDA 图 1：各列缺失值数量
fig, ax = plt.subplots(figsize=(8, 4))
missing = df_raw.isna().sum()
missing[missing > 0].plot(kind="bar", ax=ax, color="steelblue")
ax.set_title("各列缺失值数量（清洗前）")
ax.set_xlabel("列名")
ax.set_ylabel("缺失值数量")
ax.tick_params(axis="x", rotation=0)
plt.tight_layout()
plt.savefig(f"{_FIGDIR}/eda_missing.png", dpi=100)
plt.close(fig)
````

````python
# EDA 图 2：数值列分布（直方图）
num_cols = ["age", "x1", "x2", "x3"]
fig, axes = plt.subplots(1, len(num_cols), figsize=(14, 3))
for ax, col in zip(axes, num_cols):
    df_raw[col].dropna().hist(ax=ax, bins=20, color="teal", alpha=0.7)
    ax.set_title(col)
    ax.set_xlabel("值")
    ax.set_ylabel("频次")
fig.suptitle("数值列分布（清洗前，忽略 NaN）", fontsize=12)
plt.tight_layout()
plt.savefig(f"{_FIGDIR}/eda_distributions.png", dpi=100)
plt.close(fig)
````
